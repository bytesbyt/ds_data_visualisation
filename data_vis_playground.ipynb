```python
# trigonometric function
X = np.arange(0,4*np.pi, 0.05)
ycos = np.cos(X)
ysin = np.sin(X)
fig, ax = plt.subplots()
ax.plot(X, ycos, 'b-', label = 'cos(x)')
ax.plot(X, ysin, 'r--', label = 'sin(x)')
ax.set_title('Trigonometric function')
ax.set_xlabel('0 to 4pi')
ax.set_ylabel('cos(x) and sin(x)')
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.set_ylim([0,20])
ax.set_xlim([-10,10])
ax.axis('scaled')
start_x = -2.5
start_y = 2
width = 5
height = 6
facecolor = 'gray'

## create a rectangle
rectangle = plt.Rectangle((start_x, start_y), width, height, fc = facecolor)
ax.add_patch(rectangle)

## add patch to it
rectangle = plt.Rectangle((-0.5, 2), 1,2, fc = 'brown')
ax.add_patch(rectangle)

## add circle patch
circle1 = plt.Circle((-1,6), 0.5, fc = 'yellow', edgecolor = 'black', lw = 1)
ax.add_patch(circle1)

circle2 = plt.Circle((1,6), 0.5, fc = 'yellow', edgecolor = 'black', lw =1)
ax.add_patch(circle2)

## add triangle to it
start_xy = np.array([-2.5, 8])
end_xy = np.array([2.5,8])
top_xy = np.array([0,10])
triangle = Polygon(np.vstack([start_xy, end_xy, top_xy]), color = 'red')
ax.add_patch(triangle)

## save the output as a jpeg
filename = 'house.jpeg'
plt.savefig(filename, format = 'jpeg')
```

```python
#import pandas library
import pandas as pd
#load the dataset
data = pd.read_csv('iris-head-num.txt')
# show the first 5 of the data
data.head()
# show the first 10 of the data
data.head(10)
# show the last 5 of the data
data.tail()
```

```python
#shows all the details about the dataset
data.describe()
#changes row to column (transpose)
data.describe().T
```

```python
# Boxplot for each numeric feature by species
plt.figure(figsize=(5,5))
plt.boxplot(data, tick_labels = ['sepallength', 'sepalwidth', 'petallength',	'petalwidth','species'])
plt.title('boxplot')
plt.tight_layout()
```

```python
# Group by species and calculate mean and standard deviation for each feature
stats = data.groupby('species').agg(['mean', 'std'])

# Define the positions for the features
features = ['sepallength', 'sepalwidth', 'petallength', 'petalwidth']
x = np.arange(len(features))

# Create the plot
fig, ax = plt.subplots()

# Loop through each species and plot a bar chart with error bars for each feature
bar_width = 0.2  # Width of the bars
for i, species_name in enumerate(data['species'].unique()):
    # Get the mean and standard deviation for each feature for the current species
    species_mean = stats.loc[species_name, (features, 'mean')]
    species_std = stats.loc[species_name, (features, 'std')]
    
    # Shift the bars for each species so that they do not overlap
    ax.bar(x + i * bar_width, species_mean, yerr=species_std, 
           width=bar_width, capsize=5, label=species_name, 
           align='center')

# Aesthetics
ax.legend(loc="upper right")
ax.set_xlabel("Features", fontsize=14, fontweight='bold')
ax.set_ylabel("Mean +/- SD", fontsize=14, fontweight='bold')
ax.set_xticks(x + bar_width)  # Adjust x-ticks to be centered with the bars
ax.set_xticklabels(features, rotation=45)
fig.tight_layout()

# Show the plot
plt.show()
```
