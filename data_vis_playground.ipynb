```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.linspace(-1,1,5)
x = np.arange(5)
y2 = np.linspace(-2,2,5)

fig = plt.figure()
ax = fig.subplots()
line1 = ax.plot(x,y)
line2 = ax.plot(x,y2)
```

```python
# having multiple plot in a figure
number_rows = 3
number_columns = 2

# create a figure
fig2 = plt.figure()
# create subplot
ax2 = fig2.subplots(number_rows, number_columns)

ax2[0,1].plot(x,y)
ax2[1,0].plot(x,y)
ax2[2,1].plot(x,y)
```
