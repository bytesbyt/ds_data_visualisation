```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.linspace(-1,1,5)
x = np.arange(5)
y2 = np.linspace(-2,2,5)

fig = plt.figure()
ax = fig.subplots()
line1 = ax.plot(x,y)
line2 = ax.plot(x,y2)
```

```python
# having multiple plot in a figure
number_rows = 3
number_columns = 2

# create a figure
fig2 = plt.figure()
# create subplot
ax2 = fig2.subplots(number_rows, number_columns)

ax2[0,1].plot(x,y)
ax2[1,0].plot(x,y)
ax2[2,1].plot(x,y)
```

```python
# basic plotting
y1 = np.arange(0,110,10)
y2 = np.random.random(11)
x1 = np.arange(11)
fig, ax = plt.subplots(2,2)
ax[0,0].plot(x1,y1)
ax[0,1].scatter(x1,y1)
ax[1,0].bar(x1,y1)
ax[1,1].barh(x1,y1)
```

```python
# trigonometric function
X = np.arange(0,4*np.pi, 0.05)
ycos = np.cos(X)
ysin = np.sin(X)
fig, ax = plt.subplots()
ax.plot(X, ycos, 'b-', label = 'cos(x)')
ax.plot(X, ysin, 'r--', label = 'sin(x)')
ax.set_title('Trigonometric function')
ax.set_xlabel('0 to 4pi')
ax.set_ylabel('cos(x) and sin(x)')
ax.legend()
```
