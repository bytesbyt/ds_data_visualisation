```python
# Ensure each column has the appropriate data type

# Convert time to datetime format
# Ensures that invalid parsing results in NaT
data['time'] = pd.to_datetime(data['time'], errors='coerce')
```

```python
# Convert updated to datetime format
# Ensures that invalid parsing results in NaT
data['updated'] = pd.to_datetime(data['time'], errors='coerce')
```

```python
#Create date column extracting only the yyyy-mm-dd from the 'time' column
data['date'] = pd.to_datetime(data['time']).dt.date

#Create date column extracting only the yyyy-mm-dd from the 'time' column
data['month'] = pd.to_datetime(data['time']).dt.month

#Create hour column extracting on the hh from the 'time' column
data['hour'] = pd.to_datetime(data['time']).dt.hour
```

```python
#count how many times earthquake happened at each hour
data.groupby(['hour']).count()
```

```python
# At which hour of the day did earthquake happened the most in between 17.12.2024-16.01.2025

data.groupby(['hour']).count()
```

```python
# calculate the mean of each column by each hour
data.groupby(['hour']).mean(numeric_only=True)
```

```python
# Convert magnitudes to a numpy array
magnitudes = np.array(data['mag'])

# create a ficure
fig = plt.figure()

# create subplot
ax = fig.subplots()

# Create a histogram (bar plot)
ax.hist(magnitudes, bins=50, color='green', edgecolor='black')

# Add title and labels
ax.set_title('Distribution of Magnitudes')
ax.set_xlabel('Magnitude')
ax.set_ylabel('Frequency')


plt.tight_layout()
```

```python
# Count occurrences of each place
place_counts = data['place'].value_counts()

# Display the top 10 places with the most earthquakes
top_places = place_counts.head(10)
print(top_places)

# Visualize the top places
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
top_places.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 Locations with the Most Earthquakes')
plt.xlabel('Place')
plt.ylabel('Number of Earthquakes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Save the updated data DataFrame to a CSV file
data.to_csv('all_month_updated.csv', index=False)
```

```python
# Create a new column 'location' with text after 'of ' from the 'place' column
data['location'] = data['place'].str.extract(r'of (.+)', expand=False)

# Display the first few rows to confirm the changes
data
```

```python
# Count occurrences of each location
location_counts = data['location'].value_counts()

# Display the top 10 locations with the most earthquakes
top_locations = location_counts.head(10)
print(top_locations)

# Visualize the top locations
plt.figure(figsize=(10, 6))
top_locations.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 Locations with the Most Earthquakes')
plt.xlabel('Location')
plt.ylabel('Number of Earthquakes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
