```python
# Create a new column 'location' with text after 'of ' from the 'place' column
data['location'] = data['place'].str.extract(r'of (.+)', expand=False)

# Display the first few rows to confirm the changes
data
```

```python
# Count occurrences of each location
location_counts = data['location'].value_counts()

# Display the top 5 locations with the most earthquakes
top_locations = location_counts.head(5)

# Custom colour palette (hex codes)
custom_colours = ['#9F7257', '#9F2B00', '#47842E', '#E7AD41', '#D3A88F', '#DF7425']

# Visualise the top locations
plt.figure(figsize=(10, 8))

# Define bar positions and width
x_positions = np.arange(len(top_locations))
bar_width = 0.7

top_locations.plot(
    kind='bar',
    color=custom_colours[:len(top_locations)], 
    edgecolor='black',
    width=bar_width
)

# Add title and labels
plt.title('Top 5 Locations with the Most Earthquakes', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Location', fontsize=12)
plt.ylabel('Number of Earthquakes', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Add data labels above each bar
for index, value in enumerate(top_locations.values):
    plt.text(index, value + 0.5, str(value), ha='center', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

```python
import random
# Group data by location and count the number of events
location_counts = data['location'].value_counts()

# Focus on the top 5 locations and combine the rest as 'Other'
top_locations = location_counts.head(5)
other_count = location_counts.iloc[5:].sum()
top_locations['Other'] = other_count

# Custom colour palette (hex codes)
custom_colours = ['#9F7257', '#9F2B00', '#47842E', '#E7AD41', '#D3A88F', '#DF7425']

# Plot the pie chart
plt.figure(figsize=(10, 8))
top_locations.plot.pie(
    autopct='%1.1f%%',  # Show percentages
    startangle=140,     # Start angle for better visualisation
    colors=custom_colours,  # Use custom colours
    fontsize=12          # Adjust font size
)

# Add title
plt.title('Proportion of Events by Location', fontsize=18, fontweight='bold')
plt.legend(title='Locations', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylabel('') 
plt.tight_layout()
plt.show()
```

```python

# Group by location and date to count events
time_series = data.groupby(['location', 'date']).size().reset_index(name='event_count')

# Focus on the top 5 most frequent locations
top_locations = data['location'].value_counts().head(5).index
filtered_data = time_series[time_series['location'].isin(top_locations)]

# Define line styles and colors
line_styles = ['-', '--', '-.', ':', '-']
# Custom colour palette (hex codes)
colors = ['#9F7257', '#9F2B00', '#47842E', '#E7AD41', '#DF7425']

# Plot time series for each location
plt.figure(figsize=(14, 8))

# Plot each location with a distinct line style and colour
for idx, location in enumerate(top_locations):
    location_data = filtered_data[filtered_data['location'] == location]
    
    # Plot raw event counts without smoothing
    plt.plot(location_data['date'], location_data['event_count'], 
             label=location, 
             linestyle=line_styles[idx % len(line_styles)], 
             color=colors[idx % len(colors)])
    
    # Annotate the highest point
    max_point = location_data.loc[location_data['event_count'].idxmax()]
    plt.text(max_point['date'], max_point['event_count'], 
             f"{int(max_point['event_count'])}", 
             fontsize=10, color=colors[idx % len(colors)])

# Add title, labels, and legend
plt.title('Event Frequency Over Time by Location', fontsize=18, fontweight='bold')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Event Count', fontsize=14)
plt.legend(title='Location', fontsize=12, loc='upper left')
plt.grid(True, alpha=0.5)

# Format the x-axis dates
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45, fontsize=10)

# Optimise layout
plt.tight_layout()
plt.show()
```
